```python
import numpy as np
import pandas as pd

data=[]
for i in range(3):
  print("Person", i+1)
  Name =input("Enter Name: ")
  Class = input("Class: ")
  rollNO = input("Roll no: ")

  data.append({
      "Name": Name,
    "Class": Class,
    "rollNO": rollNO})

df =pd.DataFrame(data)
df.to_csv("user_data.csv", index=False)
print("\nData saved successfully!")
```
